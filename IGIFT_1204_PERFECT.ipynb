```python
import re
import time
import json
import requests as r
from bs4 import BeautifulSoup as bs
import concurrent.futures
```

```python
def get_number():
    return _list.pop(0)
```

```python
def worker(category):
    urls = []
    URL = 'https://www.igift.tw/categories/' + category + "?page=%s"
#     print(url)
    counter = 1
    switch = True
    while switch:
        resp = r.get(URL%counter)
        soup = bs(resp.text, "html5lib")
        urls.extend([[x.get('href'), category] for x in soup.select('ul > li > a') if x.get('href').startswith('/products/')])
        pages = soup.select("ul.pagination > li > a")
        if pages:
            page = pages[-1].text
            if page == "Â»":
                counter += 1
            else:
                for i in range(counter+1, int(page) + 1):
                    resp = r.get(URL%i)
                    soup = bs(resp.text, "html5lib")
                    urls.extend([[x.get('href'), category] for x in soup.select('ul > li > a') if x.get('href').startswith('/products/')])
                switch = False
        else:
            switch = False
        
    
    return urls
```

```python
def data_processing(total_pd_url):
    pd_url='https://www.igift.tw'+ total_pd_url[0]
    pd_resp=r.get(pd_url)
    pd_soup = bs(pd_resp.text,'html5lib')
    pd_name = pd_soup.select_one("div > h1").text.strip()
    pd_price = pd_soup.select_one('div.price').text.strip()
    pd_description = pd_soup.select_one('div.description-container''').text.strip()

    jd = json.loads(re.findall("app.value\(\'product\', ({.*})\);", pd_resp.text)[0])
    pd_style=[]
    picture_href=[]
    for pic_url in jd['media']:
        picture_href.append(pic_url['default_image_url'])
        picture_href.append(pic_url['detail_image_url'])

    for style in jd['variations']:
        pd_style.append(style['fields'][0]["name"])

    product.append({
                'ID': 'lv'+ "{:0>6d}".format(get_number()),
                'name' : pd_name,
                'price' : pd_price,
                'category':"",
                'brand':'IGIFT',
                'memo':pd_style,
                'desc':pd_description,
                'href':pd_url,
                'style':"",
                'picture':picture_href,
                'othertsg':total_pd_url[1],
                'comment':""
            })
    print(pd_url)
```

```python
def load_to_json(product):
    with open("igift.json", "w") as f:
        jd = json.dumps(product, ensure_ascii=False, indent=4)
        jd.encode("utf-8")
        f.write(jd)
```

```python
ttime1  =time.time()
print(ttime1)
product = []
categories = []
urls_cates = []
categories_url = []
total_pd_urls = {}
_list = [i for i in range(10000)]

domain = 'https://www.igift.tw/'
categories_resp = r.get(domain)
categories_soup=bs(categories_resp.text,'html5lib')
categories_url= list(set([ x.get('href') for x in categories_soup.select('li > a') if x.get('href').startswith('/categories/')]))

ttime2  =time.time()
print(ttime2-ttime1)

for i in categories_url:
    categories.append(re.findall("/categories/(.*)",i)[0])

ttime3  =time.time()
print(ttime3-ttime2)

with concurrent.futures.ThreadPoolExecutor(max_workers=80) as executor:
    for i in executor.map(worker, categories):
        urls_cates += i
        


ttime4  =time.time()
print(ttime4-ttime3)

for urls_cate in urls_cates:
    if urls_cate[0] in total_pd_urls:
        total_pd_urls[urls_cate[0]].append(urls_cate[1])
    else: 
        total_pd_urls[urls_cate[0]] = [urls_cate[1]]
        
ttime5  =time.time()
print(ttime5-ttime4)      

with concurrent.futures.ThreadPoolExecutor(max_workers=80) as executor:
    executor.map(data_processing, list(total_pd_urls.items()))

ttime6  =time.time()
print(ttime6-ttime5) 

load_to_json(product)

ttime7 =time.time()
print('total time:' + str(ttime7-ttime1))
```
